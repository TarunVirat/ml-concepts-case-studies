```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
house_sizes = np.array([1500, 1700, 2000, 2100, 2300])  # Example feature: house sizes
house_prices = np.array([300000, 350000, 400000, 410000, 450000])  # Example target: house prices

```

```python
# Parameters
learning_rate = 0.00000001
num_iterations = 1000

# Initialize weights and bias
weight = 0.0
bias = 0.0

# Number of training examples
m = len(house_sizes)

# Gradient Descent
for i in range(num_iterations):
    # Predictions
    predictions = weight * house_sizes + bias
    
    # Calculate the error
    error = predictions - house_prices
    
    # Compute gradients
    weight_gradient = (2/m) * np.dot(house_sizes, error)
    bias_gradient = (2/m) * np.sum(error)
    
    # Update weights and bias
    weight -= learning_rate * weight_gradient
    bias -= learning_rate * bias_gradient
    
    # Print progress every 100 iterations
    if i % 100 == 0:
        cost = (1/m) * np.sum(error ** 2)
        print(f"Iteration {i}: Cost {cost}, Weight {weight}, Bias {bias}")

# Final parameters
print(f"Final Weight: {weight}, Final Bias: {bias}")

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot of the data
plt.scatter(house_sizes, house_prices, color='blue', label='Data Points')

# Line of regression
regression_line = weight * house_sizes + bias
plt.plot(house_sizes, regression_line, color='red', label='Regression Line')

# Labels and title
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($)')
plt.title('House Prices vs. Size with Regression Line')
plt.legend()

# Show plot
plt.show()
```
